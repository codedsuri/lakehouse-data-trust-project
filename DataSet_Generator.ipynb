# Product Master

```python
categories = {
    "Electronics": ["Mobile", "Laptop", "Tablet", "Accessories"],
    "Clothing": ["Shirts", "Jeans", "Shoes", "Jackets"],
    "Home Appliances": ["Kitchen", "Furniture", "Decor"],
    "Beauty": ["Skincare", "Makeup", "Haircare"],
    "Grocery": ["Snacks", "Beverages", "Essentials"],
    "Sports": ["Fitness", "Outdoor", "Gym Equipment"]
}

brands = [
    "Nike", "Samsung", "Apple", "Sony",
    "Puma", "Adidas", "Nestle", "Loreal",
    "Ikea", "Philips"
]

countries = [
    "India", "China", "USA", "Germany",
    "Vietnam", "Japan", "South Korea"
]
```

```python
data = []

for i in range(1000):
    category = random.choice(list(categories.keys()))
    sub_category = random.choice(categories[category])

    data.append({
        "product_id": f"P{i+1001}",
        "sku_id": f"SKU{i+5001}",
        "product_name": fake.word().capitalize() + " Product",
        "category": category,
        "sub_category": sub_category,
        "brand": random.choice(brands),
        "launch_date": fake.date_between(
            start_date='-5y',
            end_date='today'
        ),
        "product_status": random.choice(["Active", "Inactive"]),
        "manufacturing_cost": round(
            random.uniform(10, 5000),2
        ),
        "country_of_origin": random.choice(countries)
    })
```

```python
df_product = pd.DataFrame(data)
df_product.head()
```

```python
df_product.to_csv(
    "product_master.csv",
    index=False
)
```

# Pricing Data

```python
pricing_data = []

for i, row in df_product.iterrows():

    list_price = round(random.uniform(50, 10000), 2)

    discount_percentage = round(
        random.uniform(0, 30), 2
    )

    final_price = round(
        list_price - (
            list_price * discount_percentage / 100
        ),
        2
    )

    pricing_data.append({
        "pricing_id": f"PR{i+1001}",
        "product_id": row["product_id"],
        "list_price": list_price,
        "discount_percentage": discount_percentage,
        "final_price": final_price,
        "currency": random.choice(
            ["USD", "INR", "AUD"]
        ),
        "effective_date": fake.date_between(
            start_date='-2y',
            end_date='today'
        ),
        "pricing_status": random.choice(
            ["Active", "Expired"]
        )
    })
```

```python
# Convert to dataframe
df_pricing = pd.DataFrame(pricing_data)
```

```python
# Preview
print(df_pricing.head())
```

```python
# Row count check
print(df_pricing.shape)
```

```python
# Export CSV
df_pricing.to_csv(
    "pricing_master.csv",
    index=False
)
```

# Order Transaction

```python
df_product = pd.read_csv("product_master.csv")
df_pricing = pd.read_csv("pricing_master.csv")
df_customer = pd.read_csv("customer_master.csv")
```

## Create helper lists

```python
product_ids = df_product["product_id"].tolist()
customer_ids = df_customer["customer_id"].tolist()

pricing_dict = dict(
    zip(
        df_pricing["product_id"],
        df_pricing["final_price"]
    )
)
```

```python
# Generate Orders Data
num_orders = 100000

order_data = []

order_statuses = [
    "Delivered",
    "Cancelled",
    "Returned",
    "In Transit"
]

payment_methods = [
    "Credit Card",
    "UPI",
    "Debit Card",
    "PayPal"
]

for i in range(num_orders):

    product_id = random.choice(product_ids)
    customer_id = random.choice(customer_ids)

    quantity = random.randint(1, 5)

    unit_price = pricing_dict[product_id]

    total_order_amount = round(
        quantity * unit_price,
        2
    )

    order_data.append({
        "order_id": f"O{i+1001}",
        "customer_id": customer_id,
        "product_id": product_id,
        "order_date": fake.date_between(
            start_date='-3y',
            end_date='today'
        ),
        "quantity": quantity,
        "unit_price": unit_price,
        "total_order_amount": total_order_amount,
        "order_status": random.choice(order_statuses),
        "payment_method": random.choice(payment_methods),
        "shipping_city": fake.city(),
        "shipping_country": random.choice(
            ["India", "USA", "Australia"]
        )
    })
```

```python
# Convert to DataFrame
df_orders = pd.DataFrame(order_data)

df_orders.head()
```

```python
# Validate row count
df_orders.shape
```

```python
# Export CSV
df_orders.to_csv(
    "orders_transactions.csv",
    index=False
)
```

# Revenue Transaction

```python
# Generate Revenue Dataset
revenue_data = []

payment_statuses = [
    "Completed",
    "Pending",
    "Failed",
    "Refunded"
]

finance_systems = [
    "Stripe",
    "PayPal",
    "Bank Transfer",
    "Internal Finance System"
]

for i, row in df_orders.iterrows():

    gross_revenue = row["total_order_amount"]

    tax_amount = round(
        gross_revenue * random.uniform(0.05, 0.18),
        2
    )

    refund_amount = round(
        gross_revenue * random.uniform(0, 0.20),
        2
    ) if row["order_status"] == "Returned" else 0

    net_revenue = round(
        gross_revenue + tax_amount - refund_amount,
        2
    )

    revenue_data.append({
        "revenue_id": f"R{i+1001}",
        "order_id": row["order_id"],
        "gross_revenue": gross_revenue,
        "tax_amount": tax_amount,
        "refund_amount": refund_amount,
        "net_revenue": net_revenue,
        "transaction_date": row["order_date"],
        "payment_status": random.choice(payment_statuses),
        "finance_system": random.choice(finance_systems)
    })
```

```python
# Convert to DataFrame
df_revenue = pd.DataFrame(revenue_data)

df_revenue.head()
```

```python
# Validate row count
df_revenue.shape
```

```python
# Export CSV
df_revenue.to_csv(
    "revenue_transactions.csv",
    index=False
)
```

# Historical Transaction

```python
# Create helper lists
product_ids = df_product["product_id"].tolist()
customer_ids = df_customer["customer_id"].tolist()

pricing_dict = dict(
    zip(
        df_pricing["product_id"],
        df_pricing["final_price"]
    )
)
```

```python
# Generate Historical Dataset
num_transactions = 300000

historical_data = []

channels = [
    "Online",
    "Retail Store",
    "Mobile App"
]

statuses = [
    "Completed",
    "Returned",
    "Cancelled"
]

regions = [
    "North America",
    "Asia Pacific",
    "Europe"
]

for i in range(num_transactions):

    product_id = random.choice(product_ids)
    customer_id = random.choice(customer_ids)

    quantity = random.randint(1, 5)

    sales_amount = round(
        quantity * pricing_dict[product_id],
        2
    )

    historical_data.append({
        "transaction_id": f"H{i+1001}",
        "customer_id": customer_id,
        "product_id": product_id,
        "transaction_date": fake.date_between(
            start_date='-7y',
            end_date='-3y'
        ),
        "quantity": quantity,
        "sales_amount": sales_amount,
        "transaction_channel": random.choice(channels),
        "region": random.choice(regions),
        "transaction_status": random.choice(statuses)
    })
```

```python
# Convert to DataFrame
df_historical = pd.DataFrame(historical_data)

df_historical.head()
```

```python
# Validate row count
df_historical.shape
```

```python
# Export CSV
df_historical.to_csv(
    "historical_transactions.csv",
    index=False
)
```
